```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

x=np.arange(0,10)
y=x*x                  # or y=x**2
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x, y, color='Red', marker='o', linestyle='dashed',linewidth=1,markerfacecolor='Black', markersize=7)
plt.xlabel('X-axis', color='black')
plt.ylabel('Y-axis', color='black')
plt.title('2D-Diagram', color='black')
plt.grid()
```

```python

x=[1,2,3,4,5,6,7] #days of week
y=[160,150,140,145,175,165,180] #sales
plt.plot(x,y,color='y', marker='o')
plt.grid()
```

```python

days = [1,2,3,4,5,6,7] #days of the week 
sales_1 = [160, 150, 140, 145, 175, 165, 180] #sales of the company 1
sales_2 = [70, 90, 160, 150, 140, 145, 175] #sales of the company 2
```

```python

plt.figure(figsize=(6,5), dpi=150)
plt.plot(days, sales_1,label='sales_1',marker='o',color='r')
plt.plot(days, sales_2,label='sales_2',marker='o',color='g')
plt.title('Sales distribution',color='blue')
plt.xlabel('days of the week',color='blue')
plt.ylabel('sales',color='blue')
plt.legend(loc='upper left')
plt.grid()
```
